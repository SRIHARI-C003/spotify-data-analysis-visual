```python
plt.figure(figsize=(12,5))

plt.title("Top 20 songs with most used by the user in Trending")
sns.barplot(x=music['trackName'].value_counts().head(20).values,
            y=music['trackName'].value_counts().head(20).index,
            color='m',
            saturation=1)
plt.show()
```

```python
music_group = music.groupby(by='trackName').sum()[['key','speechiness','tempo','energy','loudness','valence']].reset_index()
music_group
```

```python
def Top_music(index,count):
    plt.subplot(2,2,index)
    plt.title("Top 10 Channels with most number of "+count)
    sns.barplot(x=music_group.sort_values(by=count, ascending=False)[count].head(10),
            y=music_group.sort_values(by=count, ascending=False)['trackName'].head(10),
            saturation=1)

plt.figure(figsize=(20,10))
Top_music(1,'key')
Top_music(2,'speechiness')
Top_music(3,'tempo')
Top_music(4,'energy')
```
