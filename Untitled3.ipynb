STANDARD SCALAR REMOVES THE MEAN AND SCALE EACH FEATURE/VARIABLE TO UNIT VARIENCE

```python
from sklearn.preprocessing import StandardScaler
```

```python
alg_1 = StandardScaler()
wine= alg_1.fit_transform(wine)
wine
```

PCA IS USED TO REMOVE THE DIMENSIONALITY OF THE DATA

```python
from sklearn.decomposition import PCA
PCA=PCA(n_components=2)
alg_2=PCA.fit_transform(wine)
PCA_wine=pd.DataFrame(alg_2,columns=['p1','p2'])
PCA_wine
```

TO GET THE ELBOW METHOD FOR OPTIMAL VALUES

WCSS= WITHIN CLUSTER SUM OF SQUARES

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,25):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(wine)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,25), wcss)
plt.title('ELBOW METHOD')
plt.show()
```

WE OBTAIN THE ELBOW FOR THE OPTIMINAL VALUES

THE PREDICTED K-MEANS VALUES ARE SHOWN BELOW THE MANUALLY

```python
kmeans = KMeans(n_clusters=4)
y=kmeans.fit_predict(wine)
print(y)
```

TO DRAW THE GRAPH WITH THE CLUSTERING IN THE DATASET

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(alg_2[:, 0], alg_2[:, 1])

ax.set_title('PCA - 2 dimensions', fontsize=15)
ax.set_xlabel('Component 1', fontsize=13)
ax.set_ylabel('Component 2', fontsize=13)

plt.show()
```

TO LABEL THE DATA WITH THE VALUE OF N CLUTERS

```python
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init=10).fit(alg_2)
labels = kmeans.labels_
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(alg_2[:, 0], alg_2[:, 1], c=labels, cmap='inferno')

ax.set_title('PCA - 2 dimensions', fontsize=15)
ax.set_xlabel('Component 1', fontsize=13)
ax.set_ylabel('Component 2', fontsize=13)

plt.show()
```

THE ABOVE GRAPTH SHOWS THE 4 DIFFERENT COLOUR CLUSTERS USING K-MEANS CLUSTERING
